```python
import pandas as pd
import numpy as np
```

```python
try:
    df = pd.read_csv("/content/datset - Sheet1.csv")  # Replace "your_sales_data.csv" with the actual filename
except FileNotFoundError:
    print("Error: Sales data file not found.")
    exit()
```

```python
df.head(5)
```

```python
# Data Cleaning
df = df.fillna(0)  # Replace NaN values with 0 or use other appropriate method
```

```python
# Data Analysis
# Sum of sales by region
sales_by_region = df.groupby("Region")["TotalSales"].sum()
print("Sum of sales by region:\n", sales_by_region)
```

```python
# Average sales per product
average_sales_per_product = df.groupby("Product")["TotalSales"].mean()
print("\nAverage sales per product:\n", average_sales_per_product)
```

```python
# Highest and lowest selling products
highest_selling_product = df.loc[df["TotalSales"].idxmax()]["Product"]
lowest_selling_product = df.loc[df["TotalSales"].idxmin()]["Product"]
print("\nHighest selling product:", highest_selling_product)
print("Lowest selling product:", lowest_selling_product)
```

```python
# NumPy calculations
sales_data = df["TotalSales"].to_numpy()
mean_sales = np.mean(sales_data)
median_sales = np.median(sales_data)
std_sales = np.std(sales_data)
```

```python
print("\nMean sales:", mean_sales)
print("Median sales:", median_sales)
print("Standard deviation of sales:", std_sales)
```
